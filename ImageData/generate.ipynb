```python

def GenerateCoordinate(shape):
    if shape == "circle":
        x = np.random.randint(60,340)
        y = np.random.randint(60,340)
    elif shape == "rect":
        x1 = np.random.randint(60,200)
        y1 = np.random.randint(3,250)

        x2 = np.random.randint(10,140)
        y2 = np.random.randint(10,140)

        x = (x1,y1);y=(x1+x2,y1+y2)
        
    elif shape == "square":
        pass 
    elif shape == "tri":
        pass 
    elif shape == "star":
        pass

    return x,y 
```
